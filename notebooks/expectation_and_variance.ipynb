### Expectation
In probability theory, expectation (also known as expected value, mean, or average) is a generalization of the weighted average concept.

#### Definition
The expectation of a random variable $X$ is defined as 

$$
\begin{aligned}
E[X] &= x_{1}p_{1} + x_{2}p_{2} + ... + x_{n}p_{n} \\
&= \sum_{i=1}^{n} x_{i}p_{i}
\end{aligned}
$$ 

where $x_{i}$ and $p_{i}$ are $i$ th a possible outcome and its probability, respectively.

#### Properties
1. $E[aX]=aE[X]$ where $a$ is a constant.
   $$
   \begin{aligned}
   E[aX] &= \sum_{i=1}^{n} ax_{i}p_{i} \\
   &= a  \sum_{i=1}^{n} x_{i}p_{i} \\
   &= a  E[X]
   \end{aligned}
   $$

2. $E[X+b]=E[X]+b$ where $b$ is a constant.  
   $$
   \begin{aligned}
   E[X+b] &= \sum_{i=1}^{n} (x_{i}+b)p_{i} \\
   &= \sum_{i=1}^{n} (x_{i}p_{i}+bp_{i}) \\
   &= \sum_{i=1}^{n} x_{i}p_{i} + \sum_{i=1}^{n} bp_{i} \\
   &= E[X] + b \sum_{i=1}^{n} p_{i} \\
   &= E[X] + b
   \end{aligned}
   $$   
   In the definition of the expectation, we make use of the property that the sum of probabilities of all possible outcomes of a random variable is equal to 1, i.e., $\sum_{i=1}^{n} p_{i}=1$.

3. $E[X+Y]=E[X]+E[Y]$    
   $$
   \begin{aligned}
   E[X+Y] &= \sum_{i=1}^{n} (x_{i} + y_{i})p_{i} \\
   &= \sum_{i=1}^{n} (x_{i}p_{i} + y_{i}p_{i}) \\
   &= \sum_{i=1}^{n} x_{i}p_{i} + \sum_{i=1}^{n}y_{i}p_{i} \\
   &= E[X] + E[Y]
   \end{aligned}
   $$

### Variance
Variance is a measure of dispersion that quantifies how far a set of numbers is spread out from their mean value.

#### Definition
The variance of a random variable $X$ is defined as the expected value of the squared deviation from the mean of $X$.
$$
\begin{aligned}
Var[X] &= E[(X-E[X])^2]
\end{aligned}
$$

#### Properties
1. $Var[X]=E[X^{2}]-E[X]^2$
   $$
   \begin{aligned}
   Var[X] &= E[(X-E[X])^2] \\
   &= E[X^{2}-2XE[X]+E[X]^2] \\
   &= E[X^{2}]-2E[X]E[X]+E[X]^2 \\
   &= E[X^2] - 2E[X]^2 + E[X]^2 \\
   &= E[X^2] - E[X]^2
   \end{aligned}
   $$

2. $Var[aX]=a^{2}Var[X]$ where $a$ is a constant value.  
   $$
   \begin{aligned}
   V[aX] &= E[(aX)^2] - E[aX]^2 \\
   &= E[a^{2}X^{2}] - (aE[X])^2 \\
   &= a^{2}E[X^{2}] - a^{2}E[X]^2 \\
   &= a^{2}(E[X^{2}]-E[X]^2) \\
   &= a^{2} Var[X]
   \end{aligned}
   $$  

3. $Var[X+b]=Var[X]$ where $b$ is a constant value.  
   $$
   \begin{aligned}
   V[X+b] &= E[(X+b)^2] - E[X+b]^2 \\
   &= E[X^{2}+2bX+b^{2}] - (E[X]+b)^2 \\
   &= E[X^{2}]+2bE[X]+b^{2}-(E[X]^{2}+2bE[X]+b^{2}) \\
   &= E[X^2] - E[X]^2 \\
   &= Var[X]
   \end{aligned}
   $$  


4. $Var[X+Y]=Var[X]+Var[Y]$    
   $$
   \begin{aligned}
   V[X+Y] &= E[(X+Y)^2] - E[X+Y]^2 \\
   &= E[X^{2}+2XY+Y^{2}] - (E[X]+E[Y])^2 \\
   &= E[X^{2}]+2E[X]E[Y]+E[Y]^{2}-(E[X]^{2}+2E[X]E[Y]+E[Y]^{2}) \\
   &= E[X^2] + E[Y^{2}] - E[X]^{2} - E[Y]^{2} \\
   &= (E[X^2] - E[X]^{2}) + (E[Y^2] - E[Y]^{2}) \\
   &= Var[X] + Var[Y]
   \end{aligned}
   $$
